# Consumer Demand Optimization and Econometric Estimation

This project analyzes consumer behavior using economic theory and econometric analysis.

The objective is to derive the consumer demand function using utility maximization and then empirically estimate demand using regression methods.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

## Economic Theory

Consumers maximize utility subject to a budget constraint.

Utility function:

U(x,y) = x^0.5 y^0.5

Budget constraint:

Px x + Py y = I

Using the Lagrangian method, we can derive the optimal consumption bundle and demand functions.

```python
price = [10,12,14,15,18,20,22,25,27,30]
income = [200,210,220,230,240,250,260,270,280,300]
quantity = [50,48,45,43,40,38,36,33,31,28]

data = pd.DataFrame({
    "Price": price,
    "Income": income,
    "Quantity": quantity
})

data
```

```python
plt.figure()

plt.scatter(data["Price"], data["Quantity"])

plt.xlabel("Price")
plt.ylabel("Quantity Demanded")
plt.title("Demand Curve")

plt.show()
```

```python
X = data[["Price","Income"]]
Y = data["Quantity"]

X = sm.add_constant(X)

model = sm.OLS(Y,X).fit()

print(model.summary())
```

## Econometric Interpretation

The regression results show the relationship between price, income, and quantity demanded.

The negative coefficient on price reflects the law of demand: as price increases, quantity demanded decreases.

Income has a positive effect on demand, indicating that the good behaves as a normal good.
